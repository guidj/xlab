```python
import pandas as pd
import random
import numpy as np
```

```python
np.random.seed(1872)
```

```python
class Object(object):
    pass

df = Object()
df.orig = Object()
df.orig.train = pd.read_csv("train.csv")
df.orig.test = pd.read_csv("test.csv")
```

```python
df.num = Object()
df.num.train = df.orig.train.copy()
df.num.test = df.orig.test.copy()

df.num.train.drop('id', axis=1, inplace=True)
df.num.test.drop('id', axis=1, inplace=True)
```

```python
for col in [x for x in df.orig.train.keys() if 'cat' in x]:
    mapper = {}
    
    # check if there there are values of a categorical var missing from the training set that exists in the test set
    
    missing_values = [x for x in df.orig.test[col].unique() if x not in df.orig.train[col].unique()]
    if any(missing_values):
        print('Column {} has missing values in training set:'.format(col))
        print('Missing values: {}'.format(missing_values))
        
    possible_values = set(list(df.orig.test[col].unique()) + list(df.orig.train[col].unique()))
    
    for i, val in enumerate(possible_values):
        mapper[val] = i
    
    df.num.train[col].replace(mapper, inplace=True)
    df.num.test[col].replace(mapper, inplace=True)
```

```python
df.num.train.describe()
```

```python
df.num.test.describe()
```

```python
# train_index = []

df.index = Object()
df.index.train = []
df.index.test = []

for i in range(0, len(df.num.train)):
    if np.random.binomial(1, 0.75, 1):
        df.index.train.append(i)
    else:
        df.index.test.append(i)

print('Training: {}\nTesting: {}'.format(len(df.index.train)/len(df.num.train), len(df.index.test)/len(df.num.train)))
```

```python
df.data = Object()
df.data.validation = df.num.test.copy()
df.data.train = df.num.train.iloc[df.index.train,:]
df.data.test = df.num.train.iloc[df.index.test,:]
```

```python
#PCA
from sklearn.decomposition import PCA

training_matrix = df.data.train.as_matrix()
X = training_matrix[:,:-1]
y = training_matrix[:,-1:]

pca = PCA()
pca.fit(X)
Xpca = pca.transform(X)
```

```python
pca.explained_variance_ratio_[:6]
```

```python
pca.components_
```

```python
n_pca_components = 6
```

```python
# TODO: build model with original data, scaled
# TODO: analyse categorical variables
```

```python
from sklearn.neural_network import MLPRegressor as NN

# nn = NN(hidden_layer_sizes=(100,100,100))
#activation='logistic', learning_rate='adaptive',
nn = NN(hidden_layer_sizes=(200, 400, 400, 200), activation='logistic', learning_rate_init=0.009, max_iter=1000)
```

```python
nn.fit(Xpca[:,:n_pca_components], y.ravel())
```

```python
nn.score(Xpca[:,:n_pca_components], y.ravel())
# nn.score(Xpca, y)
```

```python
testing_matrix = df.data.test.as_matrix()
test_X = testing_matrix[:,:-1]
test_y = testing_matrix[:,-1:]
_pca = PCA()
_pca.fit(test_X)
test_pca = _pca.transform(test_X)
```

```python
nn.score(test_pca[:,:n_pca_components], test_y.ravel())
```

```python
validation_matrix = df.data.validation.as_matrix()
validation_X = validation_matrix[:,:]
_pca = PCA()
_pca.fit(validation_X)
validation_pca = _pca.transform(validation_X)

predictions = nn.predict(validation_pca[:,:n_pca_components])
```
